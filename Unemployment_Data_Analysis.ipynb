```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
unemp = pd.read_csv('/Users/ibtisam/Downloads/(CLEAN) unemployment_rate_data - combined_data.csv')
unemp.head()
```

```python
unemp.info()
```

```python
# Converting date strings to datetime objects (in British datetime)

unemp['Date'] = pd.to_datetime(unemp['Date'], format="%d-%m-%Y")
unemp.describe()
```

```python
plt.figure(figsize=(15,5))
plt.plot(unemp['Date'], unemp['unrate'])
```

```python
# unemployment rate for women from 1948 to 2023

plt.figure(figsize=(15,5))
sns.lineplot(data=unemp, x='Date', y='unrate_women', color='red')
```

```python
# unemployment rate for men from 1948 to 2023

plt.figure(figsize=(15,5))
sns.lineplot(data=unemp, x='Date', y='unrate_men', color='purple')
```

## Unemployment rates from 1980 to 1990

```python
# subset of the original dataframe

plt.figure(figsize=(15,5))
sns.lineplot(data=unemp[(unemp.Date >= '1/1/1980')&(unemp.Date <= '12/1/1984')], x='Date', y='unrate_women')
```

```python
# unemployment rate of women between Jan 1982 and Dec 1983
# with focus on the areas where unemployment peaks where most prominent in above line graph
# below graph illustrates spikes in Jan 1982 and Jan & Feb 1983

plt.figure(figsize=(15,5))
sns.lineplot(data=unemp[(unemp.Date >= '1/1/1982')&(unemp.Date <= '1/1/1984')], x='Date', y='unrate_women', color='brown')
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=unemp[(unemp.Date >= '1/1/1980')&(unemp.Date <= '12/1/1984')], x='Date', y='unrate_men')
plt.xticks(rotation = 30);
```

```python
# unemployment rate increased for men and women around 1982
# while men having the highest unemployment rate at around 11.2%
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=unemp[(unemp.Date>='1/1/1981')&(unemp.Date<='1/1/1984')],x="Date",y="unrate_16_to_17",label="unrate_16_to_17")
sns.lineplot(data=unemp[(unemp.Date>='1/1/1981')&(unemp.Date<='1/1/1984')],x="Date",y="unrate_18_to_19",label="unrate_18_to_19")
sns.lineplot(data=unemp[(unemp.Date>='1/1/1981')&(unemp.Date<='1/1/1984')],x="Date",y="unrate_20_to_24",label="unrate_20_to_24")
sns.lineplot(data=unemp[(unemp.Date>='1/1/1981')&(unemp.Date<='1/1/1984')],x="Date",y="unrate_25_to_34",label="unrate_25_to_34")
sns.lineplot(data=unemp[(unemp.Date>='1/1/1981')&(unemp.Date<='1/1/1984')],x="Date",y="unrate_35_to_44",label="unrate_35_to_44")
sns.lineplot(data=unemp[(unemp.Date>='1/1/1981')&(unemp.Date<='1/1/1984')],x="Date",y="unrate_45_to_54",label="unrate_45_to_54")
sns.lineplot(data=unemp[(unemp.Date>='1/1/1981')&(unemp.Date<='1/1/1984')],x="Date",y="unrate_55_over",label="unrate_55_over")

```

```python
# Concluding remarks:
# Young people between the ages below the age of 20 appear to be most
# affected during the unemployment period of the early 1980s
# in comparison to people above 20 years of age.
# Spike at beginning of 1983 for those between 16-17 years old.
# Spikes for both '16 to 17' and '18 to 19' age groups during May 1983
```

## Analysing unemployment during the recession 2010

```python
recession = unemp[(unemp.Date >= '01-01-2009') &(unemp.Date <= '12-30-2014')]
recession.tail()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data = recession, x='Date', y='unrate')
```

```python
# clear spike at the end of 2009 with the unemployment rate in
# in the US nearing 10%
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data = recession, x='Date', y='unrate_men', label='men');
plt.legend
```

```python
recession['unrate_men'].describe()
```

```python
# average unemployment rate for men during the 2010 recession = 8.7%
# max = 11.1%
# min = 5.6%
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data = recession, x='Date', y='unrate_women', label='women', color='brown');
plt.legend
```

```python
recession['unrate_women'].describe()
```

```python
# average unemployment rate for women during the 2010 recession = 7.7%
# max = 9.0%
# min = 5.4%
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=recession,x="Date",y="unrate_16_to_17",label="unrate_16_to_17")
sns.lineplot(data=recession,x="Date",y="unrate_18_to_19",label="unrate_18_to_19")
sns.lineplot(data=recession,x="Date",y="unrate_20_to_24",label="unrate_20_to_24")
sns.lineplot(data=recession,x="Date",y="unrate_25_to_34",label="unrate_25_to_34")
sns.lineplot(data=recession,x="Date",y="unrate_35_to_44",label="unrate_35_to_44")
sns.lineplot(data=recession,x="Date",y="unrate_45_to_54",label="unrate_45_to_54")
sns.lineplot(data=recession,x="Date",y="unrate_55_over",label="unrate_55_over")
```

```python
# Unemployment rate reaches around 33% during the end of 2010 for the
# work eligible youth.
# The second age group most affected were the 18 to 19 followed by
# 20 to 24.
```

## Covid 19 Unemployment

```python
covid_19 = unemp[(unemp.Date>='01/01/2020')]
covid_19.head()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data = covid_19, x='Date', y='unrate_men', label='men', color='cyan');
plt.legend
```

```python
# Around March and April 2020 there was a huge spike in unemployment
# rates in the US for men which crossed over 13%.
# Between the periods of late July and October it began decreasing
```

```python
covid_19['unrate_men'].describe()
```

```python
# Most men lost their jobs at around march 2020 with unemploymment rates 
# peaking at 13.5%
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data = covid_19, x='Date', y='unrate_women', label='women', color='green');
plt.legend
```

```python
# Huge spike again for women between March and April 2020 when the 
# cases began to rise in the US.
# For women unemployment rate was higher in comparison to men, with 
# the unemployment rate at 16%.
```

```python
covid_19['unrate_women'].describe()
```

```python
# Max = 16.2%
# Average = 5.7%
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=covid_19,x="Date",y="unrate_16_to_17",label="unrate_16_to_17")
sns.lineplot(data=covid_19,x="Date",y="unrate_18_to_19",label="unrate_18_to_19")
sns.lineplot(data=covid_19,x="Date",y="unrate_20_to_24",label="unrate_20_to_24")
sns.lineplot(data=covid_19,x="Date",y="unrate_25_to_34",label="unrate_25_to_34")
sns.lineplot(data=covid_19,x="Date",y="unrate_35_to_44",label="unrate_35_to_44")
sns.lineplot(data=covid_19,x="Date",y="unrate_45_to_54",label="unrate_45_to_54")
sns.lineplot(data=covid_19,x="Date",y="unrate_55_over",label="unrate_55_over")
plt.ylabel("Unemployment rate")
```

```python
# 18 to 19 year olds were most affected around March and April 2020 with
# the unemployement rate peaking at around 34%.
# This is followed by the 16 to 17 age group with an unemployment rate
# of around 32%
```

```python
# Conlcuding remarks:
# Most women lost their jobs in March 2020, with unemployments rates
# peaking at 16.2%
# So from the above data, it evident that women were most affected 
# by the pandemic
```
